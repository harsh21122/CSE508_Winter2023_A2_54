<h1>Next, compute the nDCG (normalized discounted cumulative gain) for the dataset. This involves
calculating nDCG at position 50 and for the entire dataset.

```python
df_50 = df_filtered[:50].copy()
```

```python
df_50
```

```python
def calculate_dcg(score):
  sum = score[0]
  pos = 2
  for i in score[1:]:
    sum = sum + (i/np.log2(pos))
    pos=pos+1
  return sum
```

```python
score = df_50.iloc[:,0].values
dcg_fifty = calculate_dcg(score)
dcg_fifty
```

```python
score = df_sorted.iloc[:,0].values
i_dcg_50 = calculate_dcg(score)
i_dcg_50
```

```python
#ndcg @50
dcg_fifty/i_dcg_50
```

```python
score = df_filtered.iloc[:,0].values
dcg_entire = calculate_dcg(score)
dcg_entire
```

```python
score = df_sorted.iloc[:,0].values
i_dcg_entire = calculate_dcg(score)
i_dcg_entire
```

```python
#ndcg@entire dataset
dcg_entire/i_dcg_entire
```

<h1>For the third objective, assume a model that ranks URLs based on the value of feature 75,
which represents the sum of TF-IDF on the whole document. URLs with higher feature 75
values are considered more relevant. Any non-zero relevance judgement value is considered
relevant. Using this model, plot a Precision-Recall curve for the query "qid:4".
The curve should help visualize the trade-off between precision and recall as the model's
threshold for relevance is adjusted.

```python
df_filtered[76]
```

```python
for i in range(len(df_filtered[76])):
  value = df_filtered[76].iloc[i] 
  value = value[3:]
  df_filtered.loc[i,76] = value
df_filtered = df_filtered.astype({76:float})
```

```python
sorted_df = df_filtered.sort_values([76], ascending=[False])
sorted_df[76]
```

```python
sorted_df
```

```python
modified_relevance_list = np.array(sorted_df[0].values)
modified_relevance_list[modified_relevance_list>0]=1
relevant_documents = (modified_relevance_list==1).sum()
```

```python
pos = 1
prec = 0
rec = 0
precision_list = []
recall_list = []
for i in range(len(modified_relevance_list)):
  if modified_relevance_list[i]==1:
    prec = prec + 1
    rec = rec+1
  recall_list.append(rec/relevant_documents)
  precision_list.append(prec/pos)
  pos = pos+1
```

```python
plt.plot(recall_list,precision_list)
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

```python
index = np.argmax(precision_list[80:])
index = 80+index
print(precision_list[index])
print(recall_list[index])
plt.plot(recall_list,precision_list)
plt.scatter(recall_list[index],precision_list[index],color='red')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
